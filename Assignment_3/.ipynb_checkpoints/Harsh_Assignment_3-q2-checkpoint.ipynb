```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
frame1 = cv2.imread('video_frames/frames/rgb_image_27.png', cv2.IMREAD_GRAYSCALE)
frame2 = cv2.imread('video_frames/frames/rgb_image_28.png', cv2.IMREAD_GRAYSCALE)
frame_stack = np.stack([frame1, frame2], axis=2).astype(np.float32)
```

```python
# Calculating gradients
Iy = np.mean(np.diff(frame_stack, axis=0), axis=2)
Ix = np.mean(np.diff(frame_stack, axis=1), axis=2)

It = np.diff(frame_stack, axis=2).squeeze()

# Removing the boundary effects of differentiation
Iy = Iy[:, :-1]
Ix = Ix[:-1, :]
It = It[:-1, :-1]

# Computing magnitude and direction of gradient
gradient_magnitude = np.sqrt(Ix**2 + Iy**2)
gradient_direction = np.arctan2(Iy, Ix)

# Calculating the optical flow vectors u and v
u = -gradient_magnitude * np.cos(gradient_direction)
v = -gradient_magnitude * np.sin(gradient_direction)

k, l = 100, 120
roi_size = 20

fig, axs = plt.subplots(1, 3, figsize=(15, 5))
axs[0].set_title('Optical Flow')
axs[0].quiver(u[k:k+roi_size, l:l+roi_size], v[k:k+roi_size, l:l+roi_size], scale=25)
axs[0].set_aspect('equal', adjustable='box')
axs[1].set_title('Frame at 27')
axs[1].imshow(frame2[k:k+roi_size, l:l+roi_size], cmap='gray')
axs[2].set_title('Frame at 28')
axs[2].imshow(frame1[k:k+roi_size, l:l+roi_size], cmap='gray')

plt.show()
```
